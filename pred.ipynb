```python
import tensorflow as tf
import pandas as pd
import numpy as np
import pickle
from tensorflow.keras.models import load_model
```

```python
model=load_model('model.h5')
```

```python
## load the encoder and scaler
with open('encoders.pkl','rb') as file:
    encoders = pickle.load(file)
    label_encoder_gender = encoders['label_encoder_gender']
    one_hot_encoder_geo = encoders['one_hot_encoder_geography']
    scaler = encoders['scaler']
```

```python
# Example input data
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
```

```python
input_df=pd.DataFrame([input_data])
# Encode 'Gender'
input_df['Gender'] = label_encoder_gender.transform(input_df['Gender'])
```

```python
# One-hot encode 'Geography'
geo_encoded = one_hot_encoder_geo.transform(input_df[['Geography']])
geo_encoded_df = pd.DataFrame(geo_encoded, columns=one_hot_encoder_geo.get_feature_names_out())
```

```python
# Concatenate encoded geography with input data
input_df = pd.concat([input_df.drop('Geography', axis=1), geo_encoded_df], axis=1)
```

```python

# Scale the features
input_scaled = scaler.transform(input_df)
```

```python
# Make prediction
prediction = model.predict(input_scaled)
prediction_proba = prediction[0][0]
```

```python
print(f"Prediction Probability: {prediction_proba:.4f}")
print(f"Churn Prediction: {'Yes' if prediction_proba > 0.5 else 'No'}")
```
